```python
import cv2
```

```python
detector = MTCNN()

from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/Rachit2.jpeg')
cv2_imshow(img)
```

```python
output = detector.detect_faces(img)

print(output)
```

```python
x,y,width,height = output[0]['box']
cv2.rectangle(img,pt1=(x,y),pt2=(x+width,y+height),color=(255,0,0),thickness=3)
cv2_imshow(img)
```

```python
left_eyeX,left_eyeY = output[0]['keypoints']['left_eye']
nose_X,nose_Y = output[0]['keypoints']['nose']

cv2.circle(img,center=(nose_X,nose_Y),color=(255,0,0),thickness=1,radius=5)

cv2_imshow(img)
```
